```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# patching
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from dissertation_gui.protocols.owen.const import Type
```

```python
data_dir = Path(".").resolve().parent.parent / "data"
```

```python
df = pd.read_excel(data_dir / "RS-485" / "Owen.xlsx")
```

```python
df.head(10)
```

```python
df.rename(
    columns={
        "Наименование \nпараметра": "name",
        "Индекс": "index",
        "Hash-код (в 16-ой \ncис. счисления)": "hash_hex",
        "Формат представления \nданных": "type",
        "Характеристика": "description",
    },
    inplace=True,
)
```

```python
df = df[(~df["description"].isna()) & (~df["type"].isna())]
```

```python
df.head()
```

```python
df["name"] = df["name"].apply(lambda x: x.replace("*","").replace("(опер.)", "").strip(), )
```

```python
df.head()
```

```python
df["index"] = df["index"].apply(lambda x: -1 if x == "-" or np.isnan(x) else int(x))
```

```python
df.head()
```

```python
df["hash_hex"] = df["hash_hex"].apply(lambda x: '0x' + str(x).replace("'", ""))
```

```python
df["hash"] = df["hash_hex"].apply(lambda x: int(x, base=16))
```

```python
df.head()
```

```python
df["type"] = df["type"].apply(lambda x: Type(x))
```
