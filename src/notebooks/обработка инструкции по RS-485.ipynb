```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# patching
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from dissertation_gui.protocols.modbus import TRMModbusDataType
```

```python
data_dir = Path(".").resolve().parent.parent / "data"
```

```python
df = pd.read_excel(data_dir / "RS-485" / "Modbus.xlsx")
```

```python
df.head(5)
```

```python
df = df[~df["Назначение"].isna()]
```

```python
df.head()
```

```python
df["Назначение"] = df["Назначение"].apply(lambda s: s.replace("\n", " ").strip())
```

```python
df["Назначение"].tail(10)
```

```python
df["Адрес Modbus (hex)"].head(10)
```

```python
df["Адрес Modbus (hex)"] = df["Адрес Modbus (hex)"].str.replace(" ", "")
```

```python
df["Адрес Modbus (hex)"] = df["Адрес Modbus (hex)"].apply(lambda s: s.split("\n"))
```

```python
df.head(10)
```

```python
df["Адрес Modbus"] = df["Адрес Modbus (hex)"].apply(lambda arr: [eval(addr) for addr in arr])
```

```python
df.head(10)
```

```python
df["Количество регистров"] = df["Адрес Modbus"].apply(lambda x: len(x))
```

```python
df.head(10)
```

```python
df["Тип данных"] = df["Тип данных"].apply(lambda x: TRMModbusDataType(x))
```

```python
df.head(10)
```
