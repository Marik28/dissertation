```python
STEP = 1000 / 256
steps = list(range(256))
values = [(STEP * step) + STEP for step in steps]
```

```python
values
```

```python
df_values = pd.DataFrame()
```

```python
df_values["code"] = steps
```

```python
df_values["value"] = values
```

```python
df_values
```

```python
from collections import defaultdict
possible_values = defaultdict(lambda: [])
for _, r1 in df_values.iterrows():
    for _, r2 in df_values.iterrows():
        for _, r3 in df_values.iterrows():
            possible_value = round(((r1["value"] * r2["value"]) / (r1["value"] + r2["value"])) + r3["value"], 5)
            possible_values[possible_value] = [int(r1["code"]), int(r2["code"]), int(r3["code"])]
```

```python
possible_values
```

```python
pt_50 = [v for v in possible_values.keys() if 40 < v < 70]
```

```python
df["code"] = df["calc"].apply(lambda x: possible_values[x])
```

```python
pt_100 = [value for value in possible_values.keys() if 80 < value < 140 ]
```

```python
def create_dataframe(pt_path, values):
    df = pd.read_csv(pt_path)
    df["calc"] = df["R"].apply(lambda x: find_nearest(values, x))
    df["error"] = abs(df["R"] - df["calc"])
    df["code"] = df["calc"].apply(lambda x: possible_values[x])
    df["R1_code"] = df["code"].apply(lambda x: x[0])
    df["R2_code"] = df["code"].apply(lambda x: x[1])
    df["R3_code"] = df["code"].apply(lambda x: x[2])
    df["R1"] = df["R1_code"].apply(lambda x: STEP * x + STEP)
    df["R2"] = df["R2_code"].apply(lambda x: STEP * x + STEP)
    df["R3"] = df["R3_code"].apply(lambda x: STEP * x + STEP)
    return df
```

```python
df_pt_100 = create_dataframe("data/pt_100.csv", pt_100)
```

```python
df_pt_100
```

```python
df_pt_100.to_csv("data/recalculated/pt_100.csv", index=False)
```
